### ***Çoklu Doğrusal Regresyon***
#### **çoklu Doğrusal Regresiyon Nedir?**
● Basit doğrusal regresyondan farklı olarak, ikiden fazla değişken arasındaki ilişkiyi tanımlamak için kullanılan bir modelin yaklaşımıdır.

#### Doğrusal Regresyon (Simple Linear Regression) 
$$
y = b_0 + b_1 x
$$

#### Çoklu Doğrusal Regresyon (Multiple Linear Regression)
$$
y = b_0 + b_1 x_1 + b_2 x_2 + \dots + b_n x_n
$$

##### Çoklu Doğrusal Regresyon Model Eğitimi ve Testi
● Bu bölümde çoklu regresyon modelinin eğitimini ve testini gerçekleştireceğiz

```python
# Kütüphaneleri İçe aktaralım
import numpy as np
import pandas as pd

# Sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Veriyi içe aktar
veri = pd.read_csv("Downloads/coklu_dogrusal_regresyon_veriseti.csv", sep = ';')
veri.head()
```

```python
# deneyim ve yaşı bağımsız değişkenler olarak alacağız 
x = veri.iloc[:,[0,2]].values #iloc'ta böyle parantez içinde istedğimiz değişkenleri alabiliriz
print(x)
```

```python
# Maaş'ı ise bağımlı değişken olarak alalım.
y = veri.maas.values.reshape(-1,1)
y
```

```python
# Çoklu Doğrusal Regresiyon Modeli Eğitimi

# Çoklu Doğrusal Regresiyon Modeli
coklu_dogrusal_regresyon = LinearRegression()

# Doğrusal regresyon eğitimi
coklu_dogrusal_regresyon.fit(x, y)
```

```python
# UNUTMA!!! SKLEARN SENDEN ARRAY İSTER SÜTÜN ŞEKLİNDE ÖYLE MODELİNİ FİT YANİ EĞİTEBİLİRSİN
# SANA ARRAY VERİR. O YÜZDEN SONUÇLARI ALIRKEN İNDEX BELİRT: sonuc[0]
# TEST 1
test_verisi1 = np.array([[10,35]]) # ilk [ parantezi arrray 2. [ ise tek array olduğunu satır ve sütün değişkeninin böyle olduğunu belirttik
test_sonucu1 = coklu_dogrusal_regresyon.predict(test_verisi1)
print("10 Yıllık deneyim ve 35 yaş sunucu çıkan maaş: {} TL".format(test_sonucu1[0]))
```

```python
# TEST 2
test_verisi2 = np.array([[5, 35]])
test_sonucu2 = coklu_dogrusal_regresyon.predict(test_verisi2)
print("5 Yıllık deneyim 35 Yaş sonucu çıkan maaş {} TL".format(test_sonucu2[0]))
```
