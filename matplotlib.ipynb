### Örnek 3 (Scatter Plot)

```python
#Örnek 3
veri = {
    "a" : np.arange(50), # 0'dan 50'ye kadar sayı oluşturduk
    "c" : np.random.randint(0, 50, 50), # 0'dan 50ye arasında rastgele tam sayı
    "d" : np.random.randn(50)} # rastgele sayı oluşturduk standart sapması 50
veri["b"] = veri["a"] + (10 * np.random.randn(50)) # a'nın üstüne belli bir gürültü eklenmiş ve 50'yle çarpmışız
veri["d"] = np.abs(veri["d"]) * 100 # verinin mutlak değeri alınmış(negatif verilere karşı) 100'le çarpılmış 

#çizdirme işlemi
plt.scatter("a", "b", c="c", s = "d", data = veri)
plt.xlabel = "a sutünü verileri"
plt.ylabel = "b sütünü verileri"
plt.show()
```

### Örnek 4 (Aynı Hizada Grafikler)

```python
# Örnek 4 ^subplot fonskiyonu ile çizebiliriz.^

isimler = ["Ahmet", "Mehmet", "Ayşe", "Fatma"]
degerler = [5, 25, 50, 100]

# yan yana 3 tane figür çizeceğim için en-boy 3 katı olsun istiyorum
plt.figure(figsize = (9,3)) # ilk bir figure ile bir alan açarız

# Subplot (çubuk grafiği) ile başlıyoruz. (sub alt demek)
plt.subplot(131) # 1 satır 3 sütünluk bir yerin 1. lokasyonuna yerleştir diyoruz 131 yazarak
plt.bar(isimler, degerler) # değerlere değişkenleri ata. 
plt.xticks(rotation = 45) # Burada daha okunaklı olması adına döndürdüm.


plt.subplot(132) # 1 satır 3 sütünlu bu 2. tablonun lokasyonuna 2. bir grafik çiziyoruz
plt.scatter(isimler, degerler) # bu sefer bir saçılım grafiği isimlerle değerleri saçarak çizdir
plt.xticks(rotation = 45)

plt.subplot(133)
plt.plot(isimler, degerler) # bu sefer de garfik çizgisi çizelim
plt.xticks(rotation = 45)

plt.suptitle("Birçok Farklı Grafik") # çoklu grafiklerin başlığını koyduk burada

plt.show() # matplotlib lib'de kullanılan grafik/leri göster fonku
```

### Örnek 6

```python
#Örnek 6
t = np.arange(0.0, 5.0, 0.1) # 0'la 5'e 0.1lik adıamlarla gittik
# numpy'daki exponansiyel fonk. yani e üzeri fonk. 
# önce siyah çizgilerle sonra da mavi noktalrla çizdirdik
plt.plot(t,(np.exp(-t) * np.cos(2 * np.pi *t)), "k", t, (np.exp(-t) * np.cos(2 * np.pi * t)), "bo")
plt.show() #
```

#### Örnek 7

```python
#örnek 7

def f(t): 
    return np.exp(-t) * np.cos(1 * np.pi *t)

t = np.arange(0.0, 5.0, 0.1)
plt.plot(t, f(t), "k", t, f(t), "bo")
plt.axis([0, 10, -1, 1])
plt.show()
```

#### Örnek 8

```python
#örnek 8 
mu1 = 100 # Ortalama
sigma1  = 15 # Standart Sapma
# Ortalamsı 0, standart sapması 1 olack şekilde rastgele sayılar üretiyor bunu standart sapma ile
# çarpıp mu1 ile toplayarak: ortalaması mu1 standart sapması sigma1 olan bir dağılım elde ediyoruz
x = mu1 + (sigma1 * np.random.randn(10000)) #10000 adet nokta var

mu2 = 110
sigma2 = 10
y = mu2 + sigma2 * np.random.randn(10000)

# Verilerin histogramı
n, bins, patches = plt.hist(x, 100, density = 1, facecolor = 'b', alpha = 0.25) # Veriler iç içe geçeceği için saydam yaptık alpha ile
n, bins, patches = plt.hist(y, 100, density = 1, facecolor = 'g', alpha = 0.25)

plt.xlabel = "Veriler"
plt.ylabel = "Olasılıklar"
# Başa r demek farklı bir yazı şekli kullanılacak demek $\ farklı bir karakter gelecek anlamı var mu karakteri gelecek.
# $(Dolar işaretiyle) latek tarzı yazış bitti şeklinde bilgilendirdik ve $ ile kapadık
plt.title(r'Verilerin Histporamı: $\mu_1 = 100, \ \mu_2 = 110$') 

# Önemli bir yere ok koymak istiyorum, hem de okun izinden veya grafiğie bir takım şeyler yazdırmak istiyorum
# Bunu Matplotlib.plt'deki annotate fonk. ile yapabilirsin
# annotate okun gösterlilmseini, yazı yazılmasını sağlıyor
plt.annotate('Gauss eğrsinin tepesi', xy=(110, 0.04), xytext = (130, 0.045), #xy okun ucu, xytext ise okun ucundan itibaren metnin sonuna kadar
        arrowprops = dict(facecolor = 'black', shrink = 0.1)) # Shrink okun kalınlık
#arrowprops(arrow proparties) okun özelliklerini gösgteren fonk.

# plotun içine bir şeyler yazdırmak istiyorum
plt.text(75, 0.025, r'$\mu_1 = 100,\ \sigma_1 = 15$')
plt.text(120, 0.035, r'$\mu_1 = 110,\ \sigma_2 = 10$')
plt.axis([40, 160, 0, 0.05]) # yatay düşey eksenler
plt.grid(True)
plt.show()
```

### Grafik hakkındaki yorumum

Standart Sapma yaparak histogram bulduk, sigma(σ) 3 aldık. Kodtaki hesaplamaları anlamadıysanız lütfen "Empirik Kural / 68-95-99.7 Kuralı" bknz.
Ayrıca alfa sigma gibi matematiksel işaretlerin nasıl yazılacağını lütfen kaynaklardan inceleyiniz.

```python
x = np.arange(0, 10, 0.01)
y = np.exp(x) # e üzere x e^^x, logaritmik ağırlandıracak, yazacak

plt.figure() #Yeni bir boş figür(grafik penceresi)

# Grafikler bazen boyurları birbirinden farklı olabilir bazen çok daha küçük sayılarla başlayıp farklı değerlerle küçük bir alana sığabielceken bazen grafikler büyük olabilir
# Bu yüzden grafikleri scale eden fonk.lar vardır onları kullan eksenleri ölçeklendirir bunlar.

# Lineer ölçekte çizdircez
plt.subplot(121)
plt.plot(x, y) # Çızık
plt.yscale('linear')
plt.title('Lineer (Doğrusal)')
plt.grid(True)

# Logaritmik Ölçekte Çizdirecez
plt.subplot(122)
plt.plot(x, y)
plt.yscale('log')
plt.title('Logaritmik')
plt.grid(True)
```
