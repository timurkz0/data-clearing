### Seaborn ile görselleştirme

Bu görselleştirmede temizlemiş olduğumuz olimpiyatlar verisetini kullandım

```python
# Kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Temizlenmiş verileri yükliyoruz
veri = pd.read_csv('Desktop/olm_temiz.csv')
```

```python
# İlk 5 satırı okutalım
veri.head()
```

### Saçılım Grafiği

```python
def sacilimGrafikCiz():
    sns.scatterplot(x = 'boy', y = 'kilo', data = veri)
    plt.title('Bot ve Ağırlık İlişkisi - Beyaz Izgara Tema')
    plt.show()

sacilimGrafikCiz()
```

```python
sns.set_style('whitegrid') # sns'deki grid türleri örn = (darkgrid)
sacilimGrafikCiz()
```

### Saçılım Grafiği - Kategorik

```python
# Renklendirme seçenekleri ve işaretleme seçenekleri yapabiliriz.
# buradaki hue = 'madalya' verilerin renklere(kategorilere) göre dağılmasını sağlar
sns.scatterplot(x = 'boy', y = 'kilo', hue = 'madalya', data = veri)
plt.title('Madalyaya Göre Boy ve Ağırlık Dağılımı')
plt.show()
# Burada Gördüğün üzere grid olsun yukarda kategorileri gösterme olsun sns kütüphanesi bizim adımıza yaptı.
```

### Saçılım Grafiği Cinsiyete Göre

```python
sns.set_style('whitegrid')

sns.scatterplot(x = 'boy', y = 'kilo', hue = 'cinsiyet', data = veri)
plt.xlabel = 'Sporcuların Boyu'
plt.ylabel = 'Sporcuların Ağırlığı'
plt.title('Madalyalı Sporcuların Ağırlık Boy Değişimi')
plt.show()
```

```python
# Burada ekstra olarak style = 'madalya' ekledik
sns.set_style('whitegrid')

# Style = 'madalya' dediğimiz zaman: verileri aynı zamanda türlere göre belirtiyor
sns.scatterplot(x = 'boy', y = 'kilo', hue = 'cinsiyet', style = 'madalya', data = veri)
plt.xlabel = 'Boy'
plt.ylabel = 'Kilo'
plt.title('Madalya Sporcuların Ağırlık-Boy Değişimleri')
plt.show()
```

### Yaşa Göre İmleç Belirleyelim

```python
# hue ve style'nin ardından imleçlerin büyüklüklerini ifade edecek imleçlerin büyüklükleri sizes
#size = 'yas', sizes = ()
sns.set_style('whitegrid')

sns.scatterplot(x = 'boy', y = 'kilo', hue = 'cinsiyet', style = 'madalya', 
                size = 'yas', sizes = (15,200), data = veri)
plt.xlabel = ('Boy')
plt.ylabel = ('Ağırlık')
plt.title('Madalyalı Sporcuların Ağırlık-Boy Değişimi')
plt.show()
```

### Saçılım Grafiği - Doğrusal Regresiyon

```python
# sns.regplot regresiyonu doğrusal yani bir çizgi halinde veren fonk.
sns.regplot(x = 'boy', y = 'kilo',data = veri, marker = '+', scatter_kws = {'alpha' : 0.2}) # scatter_kws transparanlık
plt.title('Boy ve Ağırlık DAğılımı')
plt.show()
```

### Renk Paletleri

```python
#Set1 renk paleti
sns.scatterplot(x = 'boy', y = 'kilo', hue = 'madalya', data = veri, palette = ('Set1'))
plt.title('Madalyaya göre boy ve Ağırlık Dağılımı')
plt.show()
```

```python
# rocket renk paleti
sns.scatterplot(x = 'boy', y = 'kilo', hue = 'madalya', data = veri, palette = ('rocket'))
plt.title('Madalyaya Göre Boy - Kilo Dağılımı')
plt.show()
```

### Çizgi Grafiği

```python
sns.lineplot(x = 'boy', y = 'kilo', data = veri)
plt.title('Boy ve Kiloya Göre Çizgi Grafiği')
plt.show()
```

```python
sns.lineplot(x = 'boy', y = 'kilo', hue = 'cinsiyet', data = veri)
plt.title('Cinsiyete Göre Boy ve Kilo İlişkisi Grafiği')
plt.show()
```

### Histogram

```python
# Histogram'da nerdeki kaynağa göre, neyin ağırlığı onları vereceksin
# Disturbition Plot = displot
sns.displot(veri, x = 'kilo')
plt.ylabel = ('Frekans')
plt.title('Sporcuların Kilolarına Göre Dağılımları')
plt.show()
```

#### Cinsiyete Göre Histogram

```python
sns.displot(veri, x = 'kilo', hue = 'cinsiyet') # hue kategorilere ayırmanı sağlar
plt.ylable=('frekans')
plt.title('Cinsiyete Göre Ayrıştırılmış Histogram')
plt.show()
```

```python
# multiple = 'dodge', col = 'cinsiyet' cinsiyete göre ayrıştır
sns.displot(veri, x = 'kilo', col = 'cinsiyet', multiple = 'dodge')
plt.show()
```

```python
sns.displot(veri, x = 'kilo', col = 'madalya', multiple = 'dodge')
plt.show()
```

```python
# displot'ta sadece 1 değişken yani x verirsen histogram iki değişken xy verirsen dağ görünümlü histogram çizecektir.
# kind = tür demek (kde = kernel density estimation kısaltması istatstiksel bir terim)
sns.displot(veri, x = 'kilo', y = 'boy', kind = 'kde')
plt.xlabel = 'Ağırlık'
plt.ylabel = 'Boy'
plt.title('Ağırlık - Boy Histogramı')
plt.show()
```

# İki Boyutlu Histogram

```python
sns.displot(veri, x = 'boy', y = 'kilo', hue = 'cinsiyet', kind = 'kde')
plt.xlabel = ('boy')
plt.ylabel = ('kilo')
plt.title('Cinsiyetlere Göre Boy - Kilo Histogramı')
plt.show()
```

```python
# displot'ta xy yapmak yerine kdeplot() yapsan aynı sonuçları alabilirdin ama data = veri diye belirtmeyi unutma
# sonunda fill = True ekleyebilirsin bu içlerinin doldurulmasıya alakalı şeyler
sns.kdeplot(data = veri, x = 'boy', y = 'kilo', hue = 'cinsiyet', fill = True)
plt.xlabel = ('Boy')
plt.ylabel = ('Ağırlık')
plt.title("Cinsiyete Göre Boy - Kilo Ağırlıklandırılmış Histogram")
plt.show()
```

### Çabuk Grafiği

```python
# fonk. sns.barplot()
sns.barplot(x = 'madalya', y = 'boy', data = veri)
plt.title(" Madalyaya Göre Boy Grafikleri ")
plt.show()
```

```python
sns.barplot(x = 'madalya', y = 'boy', data = veri, hue = 'cinsiyet')
plt.title("Madalyaya Göre Cinsiyete Göre Ayrıştırımış Boy Çizgi Grafiği")
plt.show()
```

```python
# col = 'sezon' dediğimizde grafikleri çizerken sezon bazında bir dizme yapmasını istiyorum
# sns.catplot() çizdirdiğimiz için içinde kind'ı da belirtmemiz gerekiyor
# sns.barplotla catplot arasındaki fark catplot kategoriye dayalı bir görselleştirme yani birden çok bar çizebileceğin gibi strip, box fln da çizebilirsin
sns.catplot(x = 'madalya', y = 'yas', hue = 'cinsiyet', col = 'sezon', data = veri,
           kind = 'bar', height = 4, aspect = 0.7) #Kind'ı belirtmediğimizde scatter yapıyor
plt.show()
```

### Box Grafik (box box, box box)

```python
sns.boxplot(x = 'sezon', y = 'boy', data = veri)
plt.show()
```

```python
sns.boxplot(x = 'sezon', y = 'boy', data = veri, hue = 'cinsiyet', palette = 'Set3')
plt.show()
```

```python
# orinet = 'h' dediğimizde yatay çiziyor belirtmezsek varsayılan olarak dikey çiziliyor
veri_gecici = veri.loc[:,['yas', 'boy', 'kilo']]
sns.boxplot(data = veri_gecici, orient = 'h', palette = 'Set1' )
plt.show()
```

```python
# col = 'kategori': cola atatığımız sütün kaç kategoriliyse o kadar grafik çizer
sns.catplot(x = 'sezon', y = 'boy', hue = 'cinsiyet', col = 'madalya',
            data = veri, kind = 'box', height = 6, aspect = 0.7)
plt.show()
```

### Isı Haritası

```python
# veri.corr: korelasyon(birbirlierne uyum), annot = True: Değerleri içine yazdır, fmt = '.1f': Format, yazdırdığın değerlerin ondalıklı gösterimi
sns.heatmap(veri.select_dtypes(include=["number"]).corr(), annot = True, linewidths = 0.5, fmt = ".1f")
plt.show() # kod çalışmadı sözel değişkenlerin de oldupundan dolayı çıkarmak lazım onları
```

### Keman Grafiği

```python
sns.violinplot(x = 'sezon', y = 'boy', data = veri)
plt.show()
```

```python
sns.catplot(x = 'sezon', y = 'boy', hue = 'cinsiyet', col = 'madalya', data = veri, kind = 'violin',
            split = True, height = 6, aspect = 0.7)
plt.title("Cinsiyete Göre Sezon - Boy Grafiği")
plt.show()
```

### Ortaklık Grafiği

```python
# Farklı özellikler arasındaki ilişkiyi farklı görselleştirme yöntemleri kullanarak incelemeye yarayan bir grafiktir
sns.jointplot(x = 'kilo', y = 'boy', data = veri, hue = 'sezon', kind = 'kde')
plt.show()
```

### Ortaklık, Box, Histogram Bir Arada

```python
# JointGrid = Çoklu Grafik kısacası ortada kesişimler, üstte ve sağda tekil dağılımı gösterirç
# Kullanıcıya Her parçayı Kontrol etme şansı verir
g = sns.JointGrid(data = veri, x = 'kilo', y = 'boy')
g.plot_joint(sns.histplot) # ortadaki grafik (iki değişkenin birlikte dağılımı)
g.plot_marginals(sns.boxplot) # Üstte ve sağda olan grafikler (tek değişkenli dağılımlar)
plt.show()
```

### Pair Plot

```python
# pairplot sayısal değişkenlerin ikili ilişkilerini scatter plot olarak,
# her bir değişkenin tek başına dağılımını histogram olarak gösterir
sns.pairplot(veri)
plt.show()
```

```python
# Pair graph - 2
# Çok Uzun sürebiliyor çalışması
g = sns.PairGrid(veri) # sayısal değişkenler arası grid(ızgara)
g.map_upper(sns.histplot) # Üst Üçgen
g.map_lower(sns.kdeplot, fill = True) # Alt Üçgen
g.map_diag(sns.histplot, kde = True) # Köşeğen Hücrelerde Ne olsun kde ile yoğunluk eğrisi eklenir
plt.show()
```
