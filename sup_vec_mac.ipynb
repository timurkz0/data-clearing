### Destek Vektör Makinesi Nedir?
● Destek vektör Makinesi, eğitim verilerini herhangi bir noktadan en uzak olan iki sınıf arasında bir karar sınırı bulunan vektör uzayı tabanlı makine öğrenmesi yöttemi olarak tanımlanmaktadır.

● Temelleri 1963 yılında Vladimir Vapnik ve Alexey Chervonekis tarafından atılan Destek Vektör Makineleri bir gözetimli öprenme algoritmasıdır.

● 1995 yılında Vlademir Vapnik, berhard Broser ve Isablle Guyon tarafından geliştirilmiştir.

● Destek Vektör Makineleri, temel olarak iki sınıfta ait verileri birbirinden en uygun şekilde destek vektörleri ile ayırmak için kullanılır.

```python
import numpy as np
import pandas as pd
```

```python
#veri
data = pd.read_csv("Downloads/ortopedik_hastalarin_biyomekanik_ozellikleri.csv")
data.head()
```

```python
data['class'] = [1 if each == 'Abnormal' else 0 for each in data['class']]
data.head(3)
```

```python
y = data['class'].values
x_data = data.drop(['class'], axis = 1)
```

```python
# Normalizasyon yapıyoruz
x = (x_data - np.min(x_data)) / (np.max(x_data)- np.min(x_data))
```

```python
# Eğitim test bölünmesi
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 1)
```

### Destek Vektör Makinesi Algoritmasının Eğitimi ve testi

```python
from sklearn.svm import SVC # support vektor classifer

# Destek Vektör Makinesi Eğitimi
svm = SVC(random_state = 1)
svm.fit(x_train, y_train)

print("Destek Vektör Makinesi Modeli Test Doğruluğu: {}".format(svm.score(x_test, y_test)))
```
